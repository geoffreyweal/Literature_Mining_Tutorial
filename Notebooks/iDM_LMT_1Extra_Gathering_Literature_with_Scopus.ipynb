# 1.2: Gathering Scientific Papers from Scopus

Scopus is a website by Elsevier that allow the user to search the literature for relevant text, and download relavant papers on mass. To begin, open https://www.scopus.com/search/form.uri?display=basic#basic in a new web browser tab.
